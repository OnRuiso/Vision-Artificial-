```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

#Se crea una imagen en 3
imagen = np.zeros((459,363,3), np.uint8)

#Rectangul del tamano de la imagen de color azul claro
cv2.rectangle(imagen, (0,0), (362,458), (206,174,0), -1)

#Creación de la figura por medio de poligonos irregulares
pts = np.array( [[97,15], [229,15], [229,40], [308,40], [308,68], [71,68], [71,40], [97,40]], np.int32)
cv2.fillPoly(imagen, [pts], (9,0,234), 1)

pts1 = np.array( [[71,69], [71,93], [97,93], [97,146], [149,146], [149,121], [122,121], [122,93], [149,93], [149,69]], np.int32)
cv2.fillPoly(imagen, [pts1], (18,27,90), 1)

pts2 = np.array( [[71,93], [96,93], [96,146], [71,146]], np.int32)
cv2.fillPoly(imagen, [pts2], (78,159,226), 1)

pts3 = np.array( [[70,93], [44,93], [44,173], [96,173], [96,147], [70,147], [70,93]], np.int32)
cv2.fillPoly(imagen, [pts3], (18,27,90), 1)

pts4 = np.array( [[97,147], [97,199], [281,199], [281,173], [203,173], [203,148], [229,148], [229,121], [203,121], [203,69], [150,69], [150,94], [123,94], [123,121], [149,121], [149,147]], np.int32)
cv2.fillPoly(imagen, [pts4], (78,159,226), 1)

pts5 = np.array( [[203,69], [229,69], [229,122], [203,122]], np.int32)
cv2.fillPoly(imagen, [pts5], (0,0,0), 1)

pts6 = np.array( [[230,69], [255,69], [255,94], [308,94], [308,120], [334,120], [334,147], [256,147], [256,122], [230,122]], np.int32)
cv2.fillPoly(imagen, [pts6], (78,159,226), 1)

pts7 = np.array( [[203,173], [203,148], [230,148], [230,123], [255,123], [255,148], [307,148], [307,173]], np.int32)
cv2.fillPoly(imagen, [pts7], (18,27,90), 1)

pts8 = np.array( [[149,200], [149,252], [203,252], [203,227], [229,227], [229,200]], np.int32)
cv2.fillPoly(imagen, [pts8], (9,0,234), 1)

pts9 = np.array( [[123,200], [123,278], [96,278], [96,331], [70,331], [70,384], [150,384], [150,359], [203,359], [203,384], [281,384], [281,331], [255,331], [255,278], [229,278], [229,226], [203,226], [203,252], [149,252], [149,200]], np.int32)
cv2.fillPoly(imagen, [pts9], (122,68,21), 1)

pts10 = np.array( [[229,278], [202,278], [202,305], [229,305]], np.int32)
cv2.fillPoly(imagen, [pts10], (34,255,249), 1)

pts11 = np.array( [[124,278], [151,278], [151,305], [124,305]], np.int32)
cv2.fillPoly(imagen, [pts11], (34,255,249), 1)

pts12 = np.array( [[123,385], [123,437], [18,437], [18,411], [44,411], [44,385]], np.int32)
cv2.fillPoly(imagen, [pts12], (18,27,90), 1)

pts13 = np.array( [[229,385], [229,437], [334,437], [334,411], [308,411], [308,385]], np.int32)
cv2.fillPoly(imagen, [pts13], (18,27,90), 1)

pts14 = np.array( [[123,200], [71,200], [71,225], [44,225], [44,252], [18,252], [18,278], [70,278], [70,304], [95,304], [95,277], [123,277]], np.int32)
cv2.fillPoly(imagen, [pts14], (9,0,234), 1)

pts15 = np.array( [[230,226], [230,277], [256,277], [256,304], [282,304], [282,278], [334,278], [334,253], [308,253], [308,226]], np.int32)
cv2.fillPoly(imagen, [pts15], (9,0,234), 1)

pts16 = np.array( [[282,279], [334,279], [334,358], [282,358], [282,331], [256,331], [256,305], [282,305]], np.int32)
cv2.fillPoly(imagen, [pts16], (78,159,226), 1)

pts17 = np.array( [[18,279], [18,358], [69,358], [69,331], [95,331], [95,305], [69,305], [69,279]], np.int32)
cv2.fillPoly(imagen, [pts17], (78,159,226), 1)

#Añadimos descripcion de la imagen
cv2.putText(imagen, 'Super mario bros 8 bits', (3,452), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,0,0), 2)

#Visualizamos la imagen en una ventana y la guardamos
cv2.imshow("Mario", imagen)
cv2.imwrite("Mario.png", imagen)

#Histograma de la imagen en general
plt.figure(figsize=(16,16))
plt.subplot(321)
plt.title('Histograma Imágen')
plt.hist(imagen.ravel(), 256, [0, 256]) 

#Histograma de la imagen para cada componente de color BGR
plt.subplot(322)
color = ('b', 'g', 'r')
for i, col in enumerate(color):
    histogram = cv2.calcHist([imagen], [i], None, [256], [0, 256])
    plt.title('Histogramam Componentes de Color')
    plt.plot(histogram, color = col)
    plt.xlim([0,256])

#Imagen transformada a escala de grises
plt.subplot(323)
img = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
plt.title('Imágen en Escala de Grises')
plt.xticks([]), plt.yticks([])
plt.imshow(img,'gray')

#Histograma de laimagen en escala de grise
plt.subplot(324)
plt.title('Histograma Imágen en Escala de Grises')
plt.hist(img.ravel(), 256, [0, 256]) 

#Ecualizacion de la imagen en escala de grises
plt.subplot(325)
clahe = cv2.createCLAHE(clipLimit=20.0)
cl1 = clahe.apply(img)
plt.title('Imágen en Escala de Grises Ecualizada')
plt.xticks([]), plt.yticks([])
plt.imshow(cl1,'gray')

#Histograma de la imagen ecualizada
plt.subplot(326)
plt.title('Histograma Imágen Ecualizada')
plt.hist(cl1.ravel(), 256, [0, 256]) 
plt.show()

#Comparacion de histogramas imagen escala de grises y ecualizada
plt.title('Comparación Histogramas Imágen en Escala de Erises y Ecualizada')
plt.hist(img.ravel(), 256, [0, 256]) ,  
plt.hist(cl1.ravel(), 256, [0, 256])
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```
