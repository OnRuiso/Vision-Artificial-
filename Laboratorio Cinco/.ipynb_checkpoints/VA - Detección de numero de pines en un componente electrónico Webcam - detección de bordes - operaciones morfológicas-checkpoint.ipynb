##### Disponga diferentes componentes electrónicos organizados sobre una línea de detección. Utilizando una webcam enfoque los pines de cada elemento y mediante una detección de bordes cuente los pines presentes para el elemento adquirido con la cámara. Determine si el elemento tiene un menor o mayor número de pines que el requerido. Utilice detección de bordes y operaciones morfológicas para cumplir con este requerimiento.

```python
import cv2
import time
import numpy as np
from matplotlib import colors
from matplotlib import pyplot as plt
import sys
import os
#----------------------------------------------------------------------------------------------------
camara = cv2.VideoCapture(1)
kernel = np.ones((3,3), np.uint8)
#----------------------------------------------------------------------------------------------------
c=input ("Ingrese el numero de pines que solicita: ")
n=int(c)
#----------------------------------------------------------------------------------------------------
while (1):
    #----------------------------------------------------------------------------------------------------
    _,frame = camara.read()
    #----------------------------------------------------------------------------------------------------
    primera1 = frame.copy()
    
    height, width = primera1.shape[:2]
    start_row, start_col = int(height * .034), int(width * .35)
    end_row, end_col = int(height * .971), int(width * .86)
    primera = primera1[start_row:end_row , start_col:end_col]
    
    #----------------------------------------------------------------------------------------------------
    camgrises = cv2.cvtColor(primera, cv2.COLOR_BGR2GRAY)
    difuminado = cv2.GaussianBlur(camgrises, (11, 11), 0)
    #----------------------------------------------------------------------------------------------------
    ratio = 3
    kernel_size = 3
    low_threshold = 15
    difuminado_blur = cv2.blur(difuminado, (3,3))
    bordeDif = cv2.Canny(difuminado_blur, low_threshold, low_threshold*ratio, kernel_size)
    #----------------------------------------------------------------------------------------------------
    nueva = bordeDif.copy()
    #----------------------------------------------------------------------------------------------------
    height, width = nueva.shape[:2]
    start_row, start_col = int(height * .35), int(width * .55)
    end_row, end_col = int(height * .55), int(width * .75)
    cropped = nueva[start_row:end_row , start_col:end_col]
    #----------------------------------------------------------------------------------------------------
    dilatacion = cv2.dilate(cropped, kernel, iterations = 1)
    #----------------------------------------------------------------------------------------------------
    (contornos,_) = cv2.findContours(dilatacion, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cv2.drawContours(primera,contornos,-1,(0,0,255), 2)
    #----------------------------------------------------------------------------------------------------
    pines1 = len(contornos)
    pines0 = int(pines1)
    #----------------------------------------------------------------------------------------------------
       
    variable = pines0 - n
    
    if variable == 0:
        print("El numero de pines es el correcto: " .format(len(contornos)))
    elif pines0 > n:
        print("El numero de pines es incorrecto, tienes mas ",variable," de pines")
    elif pines0 < n:
        print("El numero de pines es incorrecto, tiene menos ",variable," de pines")
    elif variable < 0:
        print("El numero de pines es menor que el dispositivo ingresado")
        
    #----------------------------------------------------------------------------------------------------
    
    cv2.imshow("Camara original", frame)
    cv2.imshow("Camara Difuminada al Difuminado" , bordeDif)
    cv2.imshow("Cropped Image", cropped)
    cv2.imshow("Dilatacion",dilatacion)
    
    #----------------------------------------------------------------------------------------------------
        
    tecla = cv2.waitKey(27) & 0xFF
    if tecla == 27:
        break
    #----------------------------------------------------------------------------------------------------
camara.release()
cv2.destroyAllWindows()
```
