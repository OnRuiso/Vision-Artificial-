## DIAGRAMAS DE FLUJO DE LOS PROGRAMAS

### Diagrama de flujo del primer programa

```python
import cv2
import matplotlib
from matplotlib import colors
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.pyplot as plt

imagenD = cv2.imread('GUIA 5 PUNTO 1.png')
imgD_rgb = cv2.cvtColor(imagenD, cv2.COLOR_BGR2RGB)

plt.figure()
plt.title('Diagrama de Flujo rimer programa Guia No.5 tamaño original')

plt.xticks([]), plt.yticks([])
plt.imshow(imagenD)
plt.show()
rows6,cols6,ch6 = imagenD.shape
print("Tamaño de imagen:",rows6,"x",cols6)

scale = 2.5
img_S = cv2.resize(imgD_rgb, None,fx=1./scale, fy=1./scale, interpolation=cv2.INTER_LANCZOS4)

plt.figure()
plt.title('Diagrama de Flujo primer programa Guia No.5 Escalado a 1/2.5')
plt.xticks([]), plt.yticks([])
plt.imshow(img_S)
plt.show()
rows7,cols7,ch7 = img_S.shape
print("Tamaño de imagen:",rows7,"x",cols7)

img_VV = cv2.cvtColor(img_S, cv2.COLOR_BGR2RGB)

cv2.imshow('Resultado de Homografia', img_VV)
cv2.waitKey()
cv2.destroyAllWindows()
```
