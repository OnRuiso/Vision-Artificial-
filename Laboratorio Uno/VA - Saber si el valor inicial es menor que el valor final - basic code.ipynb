#### Utilizando Python, genere un código que permita ejecutar lo planteado por el siguiente diagrama de flujo. Lo números deben poderse cambiar por el usuario:



```python
valorinicial=input('Ingrese un número entero como valor inicial: ')
try:
    ival = int(valorinicial)
except:
    ival = 'c'
while ival == 'c':
    print('El número ingresado no es un número entero, intentelo de nuevo')
    valorinicial=input('Ingrese un número entero como valor inicial: ')
    try:
        ival = int(valorinicial)
    except:
        ival = 'c'

valorfinal=input('Ingrese un número entero como valor final: ')
try:
    ival1 = int(valorfinal)
except:
    ival1 = 'c'
while ival1 == 'c':
    print('El número ingresado no es un número entero, intentelo de nuevo')
    valorfinal=input('Ingrese un número entero como valor final: ')
    try:
        ival1 = int(valorfinal)
    except:
        ival1 = 'c'
        
vi=int(valorinicial)
vf=int(valorfinal)
    
while vi < vf:
    print(vi)
    vi = vi + 1
print('Fin')
```
