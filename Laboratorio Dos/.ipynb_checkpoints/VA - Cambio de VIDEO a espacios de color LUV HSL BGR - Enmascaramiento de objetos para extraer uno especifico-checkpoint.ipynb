```python
import cv2
import cv2 as cv
import matplotlib
from matplotlib import colors
from matplotlib import pyplot as plt
import numpy as np


c=input ("Presiona L para espacio LUV, H para HLS y B para BGR o S para salir")
n=ord(c)
while n !=83:
    if n==83:
        cv2.destroyAllWindows()
    if n == 66:
        while n!=83:
            d=input ("Presiona N para el objeto negro, A para el objeto amarillo, R para el objeto rojo o S para salir")
            e=ord(d)
            if e == 82:
                cap = cv.VideoCapture(0)
                while(1):
                    _, frame = cap.read()
                    img_bgr = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)
                    rojo_alt=np.array([28,20,120])
                    rojo_baj= np.array([130,99,243])
                    mask = cv2.inRange(frame, rojo_alt, rojo_baj)
                    res = cv.bitwise_and(img_bgr,img_bgr, mask= mask)
                    cv2.imshow('BGR',img_bgr)
                    cv2.imshow ('mask',mask)
                    cv2.imshow ('rest',res)
                    k = cv.waitKey(5) & 0xFF
                    if k == 27:
                        cv.destroyAllWindows()
                        break

            if e == 65:
                cap = cv.VideoCapture(0)
                while(1):
                    _, frame = cap.read()
                    img_bgr = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)
                    amar_alt=np.array([4,97,190])
                    amar_baj= np.array([111,231,239])
                    mask = cv2.inRange(frame, amar_alt, amar_baj)
                    res = cv.bitwise_and(img_bgr,img_bgr, mask= mask)
                    cv2.imshow('BGR',img_bgr)
                    cv2.imshow ('mask',mask)
                    cv2.imshow ('rest',res)
                    k = cv.waitKey(5) & 0xFF
                    if k == 27:
                        cv.destroyAllWindows()
                        break   

            if e == 78:
                cap = cv.VideoCapture(0)
                while(1):
                    _, frame = cap.read()
                    img_bgr = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)
                    negro_alt=np.array([15,15,15])
                    negro_baj= np.array([97,86,91])
                    mask = cv2.inRange(frame, negro_alt, negro_baj)
                    res = cv.bitwise_and(img_bgr,img_bgr, mask= mask)
                    cv2.imshow('BGR',img_bgr)
                    cv2.imshow ('mask',mask)
                    cv2.imshow ('rest',res)
                    k = cv.waitKey(5) & 0xFF
                    if k == 27:
                        cv.destroyAllWindows()
                        break
            if e==83 :
                cv2.destroyAllWindows()
                break
    if n== 76:
        while n!=ord('S'):
            d=input ("Presiona N para el objeto negro, A para el objeto amarillo, R para el objeto rojo o S para salir")
            e=ord(d)
            if e == 82:
                cap = cv.VideoCapture(0)
                while(1):
            # Take each frame
                    _, frame = cap.read()
                    img_luv = cv2.cvtColor(frame, cv2.COLOR_BGR2Luv)
                    rojo_alt=np.array([28,20,120])
                    rojo_baj= np.array([130,99,243])
                    mask = cv2.inRange(frame, rojo_alt, rojo_baj)
                    res = cv.bitwise_and(img_luv,img_luv, mask= mask)
                    cv2.imshow('luv',img_luv)
                    cv2.imshow ('mask',mask)
                    cv2.imshow ('rest',res)
                    k = cv.waitKey(5) & 0xFF
                    if k == 27:
                        cv.destroyAllWindows()
                        break
            if e == 65:
                cap = cv.VideoCapture(0)
                while(1):
                    _, frame = cap.read()
                    img_luv = cv2.cvtColor(frame, cv2.COLOR_BGR2Luv)
                    amar_alt=np.array([4,97,190])
                    amar_baj= np.array([111,231,239])
                    mask = cv2.inRange(frame, amar_alt, amar_baj)
                    res = cv.bitwise_and(img_luv,img_luv, mask= mask)
                    cv2.imshow('Luv',img_luv)
                    cv2.imshow ('mask',mask)
                    cv2.imshow ('rest',res)
                    k = cv.waitKey(5) & 0xFF
                    if k == 27:
                        cv.destroyAllWindows()
                        break   

            if e == 78:
                cap = cv.VideoCapture(0)
                while(1):
                    _, frame = cap.read()
                    img_luv = cv2.cvtColor(frame, cv2.COLOR_BGR2Luv)
                    negro_alt=np.array([15,15,15])
                    negro_baj= np.array([97,86, 91])
                    mask = cv2.inRange(frame, negro_alt, negro_baj)
                    res = cv.bitwise_and(img_luv,img_luv, mask= mask)
                    cv2.imshow('Luv',img_luv)
                    cv2.imshow ('mask',mask)
                    cv2.imshow ('rest',res)
                    k = cv.waitKey(5) & 0xFF
                    if k == 27:
                        cv.destroyAllWindows()
                        break
            if e==83 :
                cv2.destroyAllWindows()
                break
    if n== 72 :
        while n!=ord('S'):
            d=input ("Presiona N para el objeto negro, A para el objeto amarillo, R para el objeto rojo o S para salir")
            e=ord(d)
            if e == 82:
                cap = cv.VideoCapture(0)
                while(1):
            # Take each frame
                    _, frame = cap.read()
                    img_hls = cv2.cvtColor(frame, cv2.COLOR_BGR2HLS)
                    rojo_alt=np.array([28,20,120])
                    rojo_baj= np.array([130,99,243])
                    mask = cv2.inRange(frame, rojo_alt, rojo_baj)
                    res = cv.bitwise_and(img_hls,img_hls, mask= mask)
                    cv2.imshow('HLS',img_hls)
                    cv2.imshow ('mask',mask)
                    cv2.imshow ('rest',res)
                    k = cv.waitKey(5) & 0xFF
                    if k == 27:
                        cv.destroyAllWindows()
                        break

            if e == 65:
                cap = cv.VideoCapture(0)
                while(1):
                    _, frame = cap.read()
                    img_hls = cv2.cvtColor(frame, cv2.COLOR_BGR2HLS)
                    amar_alt=np.array([4,97,190])
                    amar_baj= np.array([111,231,239])
                    mask = cv2.inRange(frame, amar_alt, amar_baj)
                    res = cv.bitwise_and(img_hls,img_hls, mask= mask)
                    cv2.imshow('HLS',img_hls)
                    cv2.imshow ('mask',mask)
                    cv2.imshow ('rest',res)
                    k = cv.waitKey(5) & 0xFF
                    if k == 27:
                        cv.destroyAllWindows()
                        break   

            if e == 78:
                cap = cv.VideoCapture(0)
                while(1):
                    _, frame = cap.read()
                    img_hls = cv2.cvtColor(frame, cv2.COLOR_BGR2HLS)
                    negro_alt=np.array([15,15,15])
                    negro_baj= np.array([97,86, 91])
                    mask = cv2.inRange(frame, negro_alt, negro_baj)
                    res = cv.bitwise_and(img_hls,img_hls, mask= mask)
                    cv2.imshow('HLS',img_hls)
                    cv2.imshow ('mask',mask)
                    cv2.imshow ('rest',res)
                    k = cv.waitKey(5) & 0xFF
                    if k == 27:
                        cv.destroyAllWindows()
                        break
            if e==83 :
                cv2.destroyAllWindows()
                break
    if n==83 :
        break
        cv2.destroyAllWindows()
    else:
        c=input ("Presiona L para espacio LUV, H para HLS y B para BGR o S para salir")
        n=ord(c)
```
