```python
import cv2
import cv2 as cv
import matplotlib
from matplotlib import colors
from matplotlib import pyplot as plt
import numpy as np

img = cv2.imread('numero2.png')

img_bgr = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_luv = cv2.cvtColor(img, cv2.COLOR_BGR2Luv)
img_hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)

cv2.imwrite('Imagen Original.png', img)

cv2.imshow('RGB', img)
cv2.waitKey()
cv2.destroyAllWindows()

c=input("Presiona L para espacio LUV, H para HLS y B para BGR o S para salir")
n=ord(c)

while n !=ord('S'):
    
    if n==ord('S'):
        cv2.destroyAllWindows()
    if n == ord('B'):
        while n!=ord('S'):
            cv2.imshow('BGR',img_bgr)
            cv2.waitKey()
            cv2.destroyAllWindows()
            d=input ("Presiona F para ver la flor, M para ver la mariposa o G para el gato o S para salir")
            e=ord(d)
            if e == ord('F'):
                cv2.imshow('BGR',img_bgr)

                flor_alt=np.array([28,20,120])
                flor_baj= np.array([130,99,243])
                mask = cv2.inRange(img, flor_alt, flor_baj)
                res = cv.bitwise_and(img_bgr,img_bgr, mask= mask)
                cv2.imshow ('mask',mask)
                cv2.imshow ('rest',res)
                cv2.waitKey()
                cv2.destroyAllWindows()
            if e == ord('M'):
                cv2.imshow('BGR',img_bgr)
                mari_alt=np.array([4,97,190])
                mari_baj= np.array([111,231,239])
                mask = cv2.inRange(img, mari_alt, mari_baj)
                res = cv.bitwise_and(img_bgr,img_bgr, mask= mask)
                cv2.imshow ('mask',mask)
                cv2.imshow ('rest',res)
                cv2.waitKey()
                cv2.destroyAllWindows()
            if e == ord('G'):
                cv2.imshow('BGR',img_bgr)
                gato_alt=np.array([15,15,15])
                gato_baj= np.array([97,86, 91])
                mask = cv2.inRange(img, gato_alt, gato_baj)
                res = cv.bitwise_and(img_bgr,img_bgr, mask= mask)
                cv2.imshow ('mask',mask)
                cv2.imshow ('rest',res)
                cv2.waitKey()
                cv2.destroyAllWindows()
            if e == ord('S'):
                break
            cv2.destroyAllWindows()
    if n== ord('L'):
        while n!=ord('S'):
            cv2.imshow('Luv',img_luv)
            cv2.waitKey()
            cv2.destroyAllWindows()
            d=input ("Presiona F para ver la flor, M para ver la mariposa o G para el gato o S para salir")
            e=ord(d)
            if e == ord('F'):
                cv2.imshow('Luv',img_luv)

                flor_alt=np.array([28,20,120])
                flor_baj= np.array([130,99,243])
                mask = cv2.inRange(img, flor_alt, flor_baj)
                res = cv.bitwise_and(img_luv,img_luv, mask= mask)
                cv2.imshow ('mask',mask)
                cv2.imshow ('rest',res)
                cv2.waitKey()
                cv2.destroyAllWindows()
            if e == ord('M'):
                cv2.imshow('Luv',img_luv)
                mari_alt=np.array([4,97,190])
                mari_baj= np.array([111,231,239])
                mask = cv2.inRange(img, mari_alt, mari_baj)
                res = cv.bitwise_and(img_luv,img_luv, mask= mask)
                cv2.imshow ('mask',mask)
                cv2.imshow ('rest',res)
                cv2.waitKey()
                cv2.destroyAllWindows()
            if e == ord('G'):
                cv2.imshow('Luv',img_luv)
                gato_alt=np.array([15,15,15])
                gato_baj= np.array([97,86, 91])
                mask = cv2.inRange(img, gato_alt, gato_baj)
                res = cv.bitwise_and(img_luv,img_luv, mask= mask)
                cv2.imshow ('mask',mask)
                cv2.imshow ('rest',res)
                cv2.waitKey()
                cv2.destroyAllWindows()
            if e == ord('S'):
                break
            cv2.destroyAllWindows()
    if n== ord('H'):
        while n!=ord('S'):
            cv2.imshow('HLS',img_hls)
            cv2.waitKey()
            cv2.destroyAllWindows()
            d=input ("Presiona F para ver la flor, M para ver la mariposa, G para el gato o S para salir")
            e=ord(d)
            if e == ord('F'):
                cv2.imshow('HLS',img_hls)

                flor_alt=np.array([28,20,120])
                flor_baj= np.array([130,99,243])
                mask = cv2.inRange(img, flor_alt, flor_baj)
                res = cv.bitwise_and(img_hls,img_hls, mask= mask)
                cv2.imshow ('mask',mask)
                cv2.imshow ('rest',res)
                cv2.waitKey()
                cv2.destroyAllWindows()
            if e == ord('M'):
                cv2.imshow('HLS',img_hls)
                mari_alt=np.array([4,97,190])
                mari_baj= np.array([111,231,239])
                mask = cv2.inRange(img, mari_alt, mari_baj)
                res = cv.bitwise_and(img_hls,img_hls, mask= mask)
                cv2.imshow ('mask',mask)
                cv2.imshow ('rest',res)
                cv2.waitKey()
                cv2.destroyAllWindows()
            if e == ord('G'):
                cv2.imshow('HLS',img_hls)
                gato_alt=np.array([15,15,15])
                gato_baj= np.array([97,86, 91])
                mask = cv2.inRange(img, gato_alt, gato_baj)
                res = cv.bitwise_and(img_hls,img_hls, mask= mask)
                cv2.imshow ('mask',mask)
                cv2.imshow ('rest',res)
                cv2.waitKey()
                cv2.destroyAllWindows()
            if e == ord('S'):
                break
                cv2.destroyAllWindows()
    if n == ord('S'):
        break
        
    else:
        c=input ("Presiona L para espacio LUV, H para HLS y B para BGR o S para salir")
        n=ord(c)
```
