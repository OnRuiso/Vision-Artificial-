# Ejemplo de derivada 2D - Imagen

```python
import cv2
import matplotlib
from matplotlib import colors
from matplotlib import pyplot as plt
import numpy as np
```

## Leer imagen

```python
#img = cv2.imread('test1.jpg',0)
img = cv2.imread('lena_color_256.tif',0)


#img_rgb = cv2.imread('lena_color_256.tif',cv2.IMREAD_COLOR)
#img = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)


plt.figure()
plt.title('Gray')
plt.xticks([]), plt.yticks([])
plt.imshow(img,cmap='gray')
plt.show()

# height - width - bit per pixel
h,w = img.shape
 
# print image properties.
print ("height: " + str(h))
print ("width: " + str(w))
print("img.dtype -  Data format: ",img.dtype)
print('Maximum pixel value in this image {}'.format(img.max()))
print('Minimum pixel value in this image {}'.format(img.min()))
```

## Creacin de matriz de filtrado

```python
kernel = np.ones((2,1),np.float32)
kernel[1,0] = -1
print(kernel)

kernel1 = np.ones((1,2),np.float32)
kernel1[0,1] = -1
print(kernel1)
```

## Filtrado de imagen

```python
dst = cv2.filter2D(img,-1,kernel)

plt.figure()
plt.title('Filtered')
plt.xticks([]), plt.yticks([])
plt.imshow(dst,cmap='gray')
plt.show()

dst1 = cv2.filter2D(img,-1,kernel1)

plt.figure()
plt.title('Filtered')
plt.xticks([]), plt.yticks([])
plt.imshow(dst1,cmap='gray')
plt.show()
```

```python
img_res = np.hstack((img,dst)) #stacking images side-by-side
plt.figure()
plt.title('Original vs filtered')
plt.xticks([]), plt.yticks([])
plt.imshow(img_res,cmap='gray')
plt.show()

img_res = np.hstack((img,dst1)) #stacking images side-by-side
plt.figure()
plt.title('Original vs filtered')
plt.xticks([]), plt.yticks([])
plt.imshow(img_res,cmap='gray')
plt.show()
```
