# Ejemplo de detecci√≥n de bordes mediante Canny

```python
import cv2
import matplotlib
from matplotlib import colors
from matplotlib import pyplot as plt
import numpy as np
```

## Leer imagen

```python
#img = cv2.imread('test1.jpg',0)
img = cv2.imread('lena_color_256.tif',0)


#img_rgb = cv2.imread('lena_color_256.tif',cv2.IMREAD_COLOR)
#img = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)


plt.figure()
plt.title('Gray')
plt.xticks([]), plt.yticks([])
plt.imshow(img,cmap='gray')
plt.show()

# height - width - bit per pixel
h,w = img.shape
 
# print image properties.
print ("height: " + str(h))
print ("width: " + str(w))
print("img.dtype -  Data format: ",img.dtype)
print('Maximum pixel value in this image {}'.format(img.max()))
print('Minimum pixel value in this image {}'.format(img.min()))
```

## Canny

```python
ratio = 3
kernel_size = 3
low_threshold = 20
img_blur = cv2.blur(img, (3,3))
detected_edges = cv2.Canny(img_blur, low_threshold, low_threshold*ratio, kernel_size)
```

## Bordes de imagen

```python
plt.figure()
plt.title('Filtered')
plt.xticks([]), plt.yticks([])
plt.imshow(detected_edges,cmap='gray')
plt.show()
```

```python
img_res = np.hstack((img,detected_edges)) #stacking images side-by-side
plt.figure()
plt.title('Original vs filtered')
plt.xticks([]), plt.yticks([])
plt.imshow(img_res,cmap='gray')
plt.show()
```
